enzyme.Enzyme is mostly a metadata container

```python
from oligo_maldi.metadata import Enzyme
from oligo_maldi.experiment import Experiment
```

Define enzyme


```python
demo_enzyme = Enzyme(id=0,  # numerical identifier makes things unambiguous
                     name='T4 DNA Ligase', # text identifier is better for human readability
                     mutation='S43A', # allows tracking of differences to wild-type
                     organism='T4 Phage'  # more metadata
                     )
```

Possible to create methods to handle metadata processing

```python
demo_enzyme.get_aa_position()
```

```python
demo_enzyme.some_attribute = 'Can define arbitrary metadata'
demo_enzyme.some_attribute
```

Can add many different enzymes to experiment and have them show up in the outputs

```python
enz1 = Enzyme(id=1, name="Enz1")
enz2 = Enzyme(id=2, name="Enz2")

demo_experiment = Experiment(data_folder="/Users/mitchsyberg-olsen/github/oligo_maldi/data/experiment_1")

enz1_rows = ['A', 'B', 'G', 'H']
enz2_rows = ['C', 'D', 'E', 'F']

for well, sample in demo_experiment.samples.items():
    if well[0] in enz1_rows:
        sample.enzyme = enz1
    elif well[0] in enz2_rows:
        sample.enzyme = enz2

demo_experiment.show_sample_positions()
```

```python
# enzyme metadata will show up in experiment outputs
outfile = "/Users/mitchsyberg-olsen/github/oligo_maldi/demos/enzyme_demo_results"
demo_experiment.write_to_excel(filename=outfile)
```
